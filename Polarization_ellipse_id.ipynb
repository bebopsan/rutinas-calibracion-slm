Función para rotar elementos ópticos como retardadores y polarizadores.

```python
def Rotate( element, theta):     
    """Applies a rotation operation on any polarizing element given 
    its Jones matrix representation and a angle of rotation.
    The angle is to be given in radians.  
    
    """
    import numpy as np
    
    assert isinstance(element, np.matrix)
    assert element.shape == (2,2)
    
    rotator = np.matrix([[ np.cos(theta), np.sin(theta)],\
                         [-np.sin(theta), np.cos(theta)]])
    return rotator.transpose()*element*rotator
```

## Definición de una función que caracteriza qué tanto se parece un estado de polarización generado por un polarizador y una lámina de $\lambda/4$  a un estado de polarización deseado

Esta función sirve para ser usada en conjunto con un algorítmo de minimización que permita encontrar rápidamente los ángulos   
en los que se deben poner los instrumentos del laboratorio para obtener un estado de polarización arbitrario dado por un ángulo de inclinación  
de la elipse de  polarización $(\phi)$ y un ángulo de elipticidad $(\theta)$ dados. Usando la función ellipse_check en conjunto con un algorítmo de minimización se encuentran los ángulos de elementos ópticos que generan un estado elíptico inclinado a un ángulo  $\phi$ y con un ángulo de elipticidad de:   
$$\theta = tan^{-1}\left(\pm\frac{b}{a}\right)= \pi/4$$ Tal como se muestra en la siguiente imágen:

```python
def ellipse_gen(x0,phi, theta, give_delta = False, verbose = False):
    """" This function is meant to be used in a minimization routine. 
    The objective is to check if a rotated polarizer followed by a QWP 
    retarder returns a Jones vector that has a polarization ellipse given 
    by the parametrization angles $\phi$ and $\theta$. 
        
    Inputs:
    x0: Array, contains the rotation angles for the polarizer and the QWP: [psi, alpha] 
        psi: Rotation angle of the polarizer
        alpha: Rotation angle of the lam/4 retarder
    phi: Angle in radians between the coordinate system and the mayor axes of the ellipse
    theta: Ellipticity angle of the polarization ellipse. It is function of the ratio between 
           the minor and mayor axes: theta = arctan(b/a).
    
    Outputs:
    result: Scalar, is the result of summing the square differences of polarization angles.
            Each angle (phi_c, theta_c) is extracted from the Jones vector after propagation 
            through optical elements, and compared to the required angle. 
    delta: Optional. Is the phase difference between y and x components of the field described 
           a Jones Vector.
       
    Author: Santiago Echeverri Chacón   
    """

    [psi,alpha] = x0
    
    # Matrices of horizantal oriented polarizer and retarder to be rotated:
    
    H = np.matrix([[ 1 , 0],\
                   [ 0 , 0]], dtype = 'complex')
    QWP = np.matrix([[ 1 , 0],\
                     [ 0 , -1j]], dtype = 'complex')
    
    def Rotate( element, theta):   
        
        """Applies a rotation on any polarizing element given 
        its Jones matrix representation and a angle of rotation.
        The angle is to be given in radians. The matrix is a 2*2 matrix.  
        """
        
        assert isinstance(element, np.matrix)
        assert element.shape == (2,2)
    
        rotator = np.matrix([[ np.cos(theta), np.sin(theta)],\
                             [-np.sin(theta), np.cos(theta)]],dtype = 'complex')
        return rotator.transpose()*element*rotator
    
    M = Rotate(QWP, alpha)
    P = Rotate(H, psi)
    
    # Definition of an imput Jones vector oriented along the axis of the polarizer,
    # this way the output vector keeps a normalized intensity.
    # It may be redundant when multiplying by the polarizer but this way I can change 
    # the definition later if I want other kinds of inputs.
    
    In = np.matrix([[np.cos(psi)],[np.sin(psi)*np.exp(1j*0)]],dtype = 'complex')
        
    Out = M*P*In
    
    # Extract the phase differences and field amplitudes from "Out" and calculate phi,theta:
    
    delta1 = angle( Out[0,0] )
    delta2 = angle( Out[1,0] )
    delta = delta2 - delta1
    
    Ex = abs(Out[0,0])
    Ey = abs(Out[1,0])
        
    phi_c = 1/2*np.arctan2(2*Ex*Ey*np.cos(delta), (Ex**2 - Ey**2))
    theta_c = -1/2*np.arcsin(2*Ex*Ey*np.sin(delta)/(Ex**2 + Ey**2))
    
    result =((phi - phi_c)**2 + (theta - theta_c)**2)

    if verbose:
        print('phi_c: ', phi_c,'phi: ', phi,'theta_c: ', theta_c,'theta: ', theta )
    if give_delta:
        return result, delta
    else:
        return result
```

```python
# Definition of bounds for the minimization and aplication of a minimization procedure based on the 
# reduced Broyden-Fletcher-Goldfarb–Shanno algorithm:

bnds = ((-np.pi/2, np.pi/2),(-np.pi/2, np.pi/2))

phi, theta = 0,-np.pi/4

from scipy.optimize import fmin_l_bfgs_b
res = fmin_l_bfgs_b(ellipse_gen,[ np.pi/4, np.pi/2],fprime=None,approx_grad=1,  args = (phi,theta), pgtol=1e-05,bounds =bnds)
res
```

```python
ellipse_gen(res[0],phi, theta ,True,True)
```

## Prueba de que la función de minimización entrega posiciones en las cuales se genera una elipse correspondiente al estado:     $|\psi,\delta>=|0,\pi/2>$

```python
polarizer = Rotate(H, res[0][0])
retarder = Rotate(QWP,res[0][1])

b = (retarder*polarizer*matrix([[np.cos(pi/8)],[np.sin(pi/8)]],dtype = 'complex'))
#b = (retarder*polarizer*sy.Matrix([[sy.cos(sy.pi/8)],[sy.sin(sy.pi/8)*sy.exp(1j*0)]],dtype = 'complex').evalf()).evalf()
b = b/b[0]

print('El vector de Jones asociado a este estado es:\n {0}.'.format(b))
plot_ellipse(b)
alpha =(res[0][0]*180/pi)
theta = (res[0][1]*180/pi)
print('La diferencia entre los ángulos de los elementos es de {0} grados.'.format(alpha-theta))
print('El polarizador debe estar a {0} grados y la lámina a {1} grados.'.format(alpha,theta))
```
